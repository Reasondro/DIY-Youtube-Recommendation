```python
import pandas as pd
from datetime import datetime
def normalize_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%b %d, %Y')
    except ValueError:
        return pd.to_datetime(f'{date_str}, 2025', format='%b %d, %Y')
train['Date'] = train['Date'].apply(normalize_date)
```

```python
# Fungsi untuk mengubah Views ke angka
def clean_views(view_str):
    view_str = view_str.lower().replace(" views", "").strip()
    if 'k' in view_str:
        return float(view_str.replace('k', '')) * 1000
    elif 'm' in view_str:
        return float(view_str.replace('m', '')) * 1_000_000
    else:
        return float(view_str.replace(',', ''))
train['Views'] = train['Views'].apply(clean_views)
```
